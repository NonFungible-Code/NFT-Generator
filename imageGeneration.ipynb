```python
# Image generation

from PIL import Image

for item in traits:
    #Use appropriate folder names + trait item names
    im1 = Image.open(f'./backgrounds/{item["Background"]}.png').convert('RGBA')
    im2 = Image.open(f'./skins/{item["Skin"]}.png').convert('RGBA')
    im3 = Image.open(f'./shirts/{item["Cloth"]}.png').convert('RGBA')
    im4 = Image.open(f'./eyes/{item["Eye"]}.png').convert('RGBA')
    im5 = Image.open(f'./mouths/{item["Mouth"]}.png').convert('RGBA')
    im6 = Image.open(f'./head/{item["Head"]}.png').convert('RGBA')
   

    com1 = Image.alpha_composite(im1, im2)
    com2 = Image.alpha_composite(com1, im3)
    com3 = Image.alpha_composite(com2, im4)
    com4 = Image.alpha_composite(com3, im5)
    com5 = Image.alpha_composite(com4, im6)

    rgb_im = com5.convert('RGB')

    file_name = str(item["tokenId"]) + ".jpg"
    rgb_im.save("./generatedImages/" + file_name)
    print(f'{str(item["tokenId"])} done')
```
